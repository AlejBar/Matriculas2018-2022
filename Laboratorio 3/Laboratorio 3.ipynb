## Se puede observar de acuerdo al gráfico anterior que la cantidad existente de registros correspondientes al 2021 es muy pequeña, por lo tanto se procede a eliminar los mismos

```python
# contar registros por año de nacimiento
conteo_por_anio = df_matricula['anno_inf'].value_counts()
conteo_por_anio
```

```python
# filtrar años de interés, iniciaremos por los nacidos entre 2009 y 2020
filtro = (conteo_por_anio.index == 2019)
conteo_por_anio_filtro = conteo_por_anio.loc[filtro]
conteo_por_anio_filtro
```

```python

# Identificar los índices de los registros a eliminar
indices_eliminar = df_matricula[(df_matricula['anno_inf'] == 2019)].index

# Eliminar los registros
df_resultados_1 = df_matricula.drop(indices_eliminar)

print("Se eliminaron:", conteo_por_anio_filtro.sum(),'registros que pertencen al año 2019')
```

```python
# Definir el orden personalizado para los grados
orden_grados = {'Prejardín': 0, 'Prejardin II':1,'Jardín I': 2, 'Transición': 3,
                'Primero': 4, 'Segundo': 5, 'Tercero': 6, 'Cuarto': 7, 'Quinto': 8, 'Sexto': 9,
                'Septimo': 10, 'Octavo': 11, 'Noveno': 12, 'Decimo': 13, 'Once': 14,
                'Ciclo 1 Adultos': 15, 'Ciclo 2 Adultos': 16, 'Ciclo 3 Adultos': 17,
                'Ciclo 4 Adultos': 18, 'Ciclo 5 Adultos': 19, 'Ciclo 6 Adultos': 20,
                'Aceleración del Aprendizaj': 21}

# Ordenar la serie matricula_por_grado según el orden personalizado
matricula_por_grado_ordenado = matricula_por_grado.reindex(sorted(matricula_por_grado.index, key=lambda x: orden_grados.get(x, float('inf'))))

# Graficar el número de matrículas por grado con el orden personalizado en el eje x
plt.figure(figsize=(10, 6))
matricula_por_grado_ordenado.plot(kind='bar')
plt.title('Distribución del total de matrículas por grado')
plt.xlabel('Grado')
plt.ylabel('Total de matrícula')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Teniendo en cuenta que en **Colombia**, el sistema educativo organiza los grados escolares de acuerdo a las etapas particulares del desarrollo cognitivo, emocional y social de los estudiantes. A continuación, te proporciono información sobre los grados y las edades promedio en cada nivel:

1. **Preescolar**:
   - **Prejardín**: 3 a 5 años.
   - **Transición**: 4 a 5 años.

2. **Básica – Primaria**:
   - **1° hasta 5°**: Enfoca en la adquisición de habilidades básicas de lectura, escritura y matemáticas, además de introducir otras áreas como ciencias naturales, ciencias sociales y educación artística. La edad promedio para estos grados es de **6 a 10 años**.

3. **Básica – Secundaria**:
   - **6° hasta 9°**: Profundiza en diferentes áreas de conocimiento y fortalece las habilidades académicas. La edad promedio para estos grados es de **11 a 14 años**.

4. **Media – Bachiller**:
   - **10° hasta 11°**: En este ciclo, los estudiantes eligen áreas de enfoque según sus intereses y aptitudes, como ciencias naturales, ciencias sociales, humanidades, técnico y tecnológico, entre otros. La edad promedio para estos grados es de **15 a 17 años** [[1](https://www.mineducacion.gov.co/1621/article-82787.html)]

```python
# Convertir variables categóricas en variables numéricas
# Puedes usar técnicas como Label Encoding o One-Hot Encoding. Para este ejemplo, usaremos One-Hot Encoding.

# Primero, crearemos variables dummy para 'grado' y 'departamento'
df_dummies = pd.get_dummies(df_resultados_1['grado'])

# Luego, concatenaremos las variables dummy al DataFrame original
df_numerico = pd.concat([df_resultados_1[['edad']], df_dummies], axis=1)

# Calcular la matriz de correlación
matriz_correlacion = df_numerico.corr()

# Visualizar la matriz de correlación
plt.figure(figsize=(20, 20))
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de correlación entre edad, grado')
plt.show()
```
