```python
#Importar librerias
import os
import altair as alt
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pyodbc
import warnings
import sqlite3
from sodapy import Socrata
```

```python
# Crea una instancia de la clase Socrata con dos argumentos:
    # 1: "www.datos.gov.co": Es la URL del portal de datos de Colombia (www.datos.gov.co) al que se quiere conectar.
    # 2: None: Este argumento es utilizado para especificar un token de acceso a la API si se requiere autenticación,
        # pero en este caso se pasa como None, lo que significa que no se está usando autenticación para acceder a los datos.
client = Socrata("www.datos.gov.co", None)
# Utiliza el método get del objeto client para hacer una solicitud a la API de Socrata.
# limit=15435: Limita la cantidad de registros que se devuelven a 15,435.
results = client.get("sty3-c395", limit=1500000)
```

```python
#Crear dataframe
df_matricula_educacion = pd.DataFrame.from_records(results)
df_matricula_educacion.to_csv('df_matricula_educacion.csv', index=False)
```

```python
# Análisis de valores faltantes por columna
datos_faltantes = df_matricula_educacion.isnull().sum()
porcentajes_faltantes = (df_matricula_educacion.isnull().sum() / len(df_matricula_educacion
                                                                           )) * 100

faltantes_df = pd.DataFrame({
    'Valores faltantes': datos_faltantes,
    'Porcentaje (%)': porcentajes_faltantes
})

# Ordenar de manera descendente por porcentaje de valores faltantes
faltantes_df = faltantes_df.sort_values(by='Porcentaje (%)', ascending=False)

faltantes_df
```

```python
# Descripción de las variables categóricas

df_matricula_educacion.describe(include=['object'])
```

```python
# Descripción estadística de las variables numéricas sin notación científica

# Se configura la forma en que los números flotantes se muestran al imprimir un DataFrame en pandas.
    # Se define función lambda que toma un número x y devuelve ese número formateado con 1 decimal.
pd.set_option('display.float_format', lambda x: '%.1f' % x)

df_matricula_educacion.describe()
```

```python
# Identificar valores únicos por columna
def varlor_unico_DF():
  print("\nValores únicos por columna:")
  for column in df_matricula_educacion.columns:
      unique_values = df_matricula_educacion[column].unique()
      print(f"\n{column} ({len(unique_values)} valores únicos):\n{unique_values[:10]}")
      if len(unique_values) > 10:
          print("...")
varlor_unico_DF()
```

```python
# Iterar sobre cada columna del DataFrame y mostrar la cantidad de tipos únicos de variables
for columna in df_matricula_educacion.columns:
    print("Cantidad de tipos únicos de variables en la columna:", columna)
    print(df_matricula_educacion[columna].value_counts())
    print()
```

```python
#Identificar los valores en donde la columna departamento es nulo
df_matricula_educacion[df_matricula_educacion['departamento'].isnull()]
```

```python
# Reemplazar los valores vacíos en la columna 'departamento' por "AMAZONAS" imputar
df_matricula_educacion['departamento'].fillna("AMAZONAS", inplace=True)
```

```python
# Verificar que se hayan reemplazado correctamente columna departamento.
varlor_unico_DF()
```

```python
columnas_a_eliminiar=['cod_sector','cod_calendario','cod_zona','cod_tipo_jornada',
                      'codigo_grado','cod_caracter','cod_especialidad','cod_genero',
                      'cod_sector_conpes']
df_matricula_educacion_2 = df_matricula_educacion.drop(columns=columnas_a_eliminiar)
```

```python
# Convertir a numérico
df_matricula_educacion_2['edad'] = pd.to_numeric(df_matricula_educacion_2['edad'], errors='coerce')
plt.figure(figsize=(8, 6))
plt.boxplot(df_matricula_educacion_2['edad'])

```

```python
conteo_por_anio = df_matricula_educacion_2['edad'].value_counts()

# filtrar años de interés, iniciaremos por la edad de 0 a hasta 20
filtro = (conteo_por_anio.index >= 0) & (conteo_por_anio.index <= 20)
conteo_por_anio_filtro = conteo_por_anio.loc[filtro]

# mostrar resultados
print(conteo_por_anio_filtro)
plt.show()
print("Entre la edad de 0 hasta 20 años, hay un total de matriculados:",conteo_por_anio_filtro.sum())
```

```python
df_matricula_educacion_2[df_matricula_educacion_2['edad'] > 20]
```

```python
# Obtener los índices de los registros mayores a 20
indices_eliminar = df_matricula_educacion_2[df_matricula_educacion_2['edad'] > 20].index

# Eliminar los registros
df_resultados = df_matricula_educacion_2.drop(indices_eliminar)

```
